### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python
duongdan='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv(duongdan,sep='\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
topmon=chipo['item_name'].value_counts().head(5)
plt.bar(topmon.index,topmon.values)
plt.title('Top 5 mn c t nhiu nht')
plt.ylabel('S lng t')
plt.xticks(rotation=45)
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
donhang=chipo.groupby('order_id').agg({'item_price':'sum','quantity':'sum'})
plt.scatter(donhang['item_price'],donhang['quantity'])
plt.xlabel('Gi n hng ($)')
plt.ylabel('Tng s mn')
plt.title('S lng mn theo gi n hng')
plt.show()
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
# Cu hi: Mn no c tng doanh thu cao nht?
chipo['item_price']=chipo['item_price'].str.replace('$','').astype(float)
chipo['doanhthu']=chipo['item_price']*chipo['quantity']
doanhthutheomon=chipo.groupby('item_name')['doanhthu'].sum().sort_values(ascending=False).head(10)

plt.barh(doanhthutheomon.index,doanhthutheomon.values)
plt.xlabel('Doanh thu ($)')
plt.title('Top 10 mn c doanh thu cao nht')
plt.gca().invert_yaxis()
plt.show()
```
