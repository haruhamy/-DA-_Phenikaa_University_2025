# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic 

```python
link = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(link)
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId', inplace=True)
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
sex_count = titanic['Sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(sex_count, labels=sex_count.index, autopct='%1.1f%%', startangle=90, colors=['lightblue','pink'])
plt.title('Tỷ lệ Nam/Nữ trong dữ liệu Titanic')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
colors = {'male': 'blue', 'female': 'red'}
plt.figure(figsize=(8,6))
for gender in titanic['Sex'].unique():
    subset = titanic[titanic['Sex'] == gender]
    plt.scatter(subset['Fare'], subset['Age'], label=gender, alpha=0.5, c=colors[gender])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Scatterplot giá & tuổi (theo giới tính)')
plt.legend()
plt.show()
```

### Step 7. How many people survived?

```python
survived_count = titanic['Survived'].value_counts()
print(f"Số người sống sót: {survived_count[1]}")
```

### Step 8. Create a histogram with the Fare payed

```python
plt.figure(figsize=(8,6))
titanic['Fare'].hist(bins=20, color='skyblue')
plt.xlabel('Fare')
plt.ylabel('Số lượng hành khách')
plt.title('Phân phố giá vé trên Titanic')
plt.show()
```

### BONUS: Create your own question and answer it.

```python
# Tỷ lệ sống sót theo lớp hành khách
survived_by_class = titanic.groupby('Pclass')['Survived'].mean()
survived_by_class.plot(kind='bar', color='green')
plt.title('Tỷ lệ sống sót theo hàng lớp')
plt.xlabel('Pclass')
plt.ylabel('Tỷ lệ sống sót (%)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
