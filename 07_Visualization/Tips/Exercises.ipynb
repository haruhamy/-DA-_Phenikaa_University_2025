### Step 1. Import the necessary libraries:

```python
import pandas as pd;import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
link='https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv'
```

### Step 3. Assign it to a variable called tips

```python
tips=pd.read_csv(link)
```

```python
tips
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
```

```python
tips
```

### Step 5. Plot the total_bill column histogram

```python
tips['total_bill'].hist();plt.title('Histogram total_bill');plt.xlabel('Total Bill');plt.ylabel('Tần suất');plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips['total_bill'],tips['tip']);plt.xlabel('Total Bill');plt.ylabel('Tip');plt.title('Scatter giữa total_bill và tip');plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10);plt.xlabel('Tổng hóa đơn');plt.ylabel('Tiền tip');plt.title('Liên hệ bill - tip - size');plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.bar(tips['day'],tips['total_bill']);plt.title('Tổng bill theo ngày');plt.ylabel('Total bill');plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
import seaborn as sns;sns.scatterplot(x='tip',y='day',hue='sex',data=tips);plt.title('Scatter tip theo ngày và giới');plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x='day',y='total_bill',hue='time',data=tips);plt.title('Boxplot hóa đơn theo ngày và bữa ăn');plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig,ax=plt.subplots(1,2,figsize=(12,5));tips[tips['time']=='Dinner']['tip'].hist(ax=ax[0]);ax[0].set_title('Dinner');tips[tips['time']=='Lunch']['tip'].hist(ax=ax[1]);ax[1].set_title('Lunch');plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig,ax=plt.subplots(1,2,figsize=(12,5));sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips[tips['sex']=='Male'],ax=ax[0]);ax[0].set_title('Nam');sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips[tips['sex']=='Female'],ax=ax[1]);ax[1].set_title('Nữ');plt.show()
```
